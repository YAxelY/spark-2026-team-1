# Task 4: Risk Logic
This notebook implements the core medical logic mapping patient health metrics to a risk score based strictly on the provided clinical thresholds.


```python
def assess_risk(row):
    """
    Evaluates a patient's risk level based on clinical markers.
    Returns a tuple: (Risk Level String, Risk Score Integer, Risk Factors String)
    """
    score = 0
    reasons = []
    
    # 1. Pregnancy Number
    # Any number >= 5 is considered high risk
    if row.get('pregnancy_number', 0) >= 5:
        score += 2
        reasons.append("High pregnancies (>= 5)")
        
    # 2. Blood Pressure
    sys = row.get('systolic_bp', 0)
    dia = row.get('diastolic_bp', 0)
    if sys >= 160 or dia >= 110:
        score += 3
        reasons.append("Severe High BP")
    elif sys >= 130 or dia >= 90:
        score += 2
        reasons.append("High BP")
    elif (0 < sys < 90) or (0 < dia < 60):
        score += 1
        reasons.append("Low BP")
        
    # 3. Blood Glucose
    glu = row.get('blood_glucose_mmol', 0)
    if glu >= 7.0:
        score += 3
        reasons.append("Severe/Overt Diabetes (> 7.0)")
    elif glu >= 5.1:
        score += 2
        reasons.append("Gestational Diabetes (5.1 - 6.9)")
    elif 0 < glu < 3.9:
        score += 1
        reasons.append("Low Glucose")
        
    # 4. Haemoglobin
    hb = row.get('haemoglobin_g_dl', 0)
    if 0 < hb < 7.0:
        score += 3
        reasons.append("Severe Anemia (< 7.0)")
    elif 7.0 <= hb <= 9.9:
        score += 2
        reasons.append("Moderate Anemia (7.0 - 9.9)")
    elif hb >= 13.0:
        score += 1
        reasons.append("Elevated Haemoglobin")
        
    # 5. Antenatal Visits
    vis = row.get('antenatal_visits', 0)
    if vis >= 15:
        score += 2
        reasons.append("High Visits (>= 15)")
    elif vis < 4:
        score += 2
        reasons.append("Inadequate Visits (< 4)")
    elif vis < 8:
        score += 1
        reasons.append("Suboptimal Visits (4 - 7)")
        
    # Final Risk Classification Logic
    level = "Low"
    if score >= 5:
        level = "High"
    elif score >= 2:
        level = "Medium"
        
    return level, score, ", ".join(reasons)
```

### Testing the Risk Logic
Below is a quick test simulating a patient to ensure the logic behaves correctly.


```python
# Test patient with Severe High BP and Low Glucose
sample_patient = {
    'pregnancy_number': 2,
    'systolic_bp': 170,
    'diastolic_bp': 120,
    'blood_glucose_mmol': 3.5,
    'haemoglobin_g_dl': 11.0,
    'antenatal_visits': 10
}

risk_level, risk_score, risk_factors = assess_risk(sample_patient)

print(f"Risk Level: {risk_level}")
print(f"Total Score: {risk_score}")
print(f"Flagged Factors: {risk_factors}")
```
